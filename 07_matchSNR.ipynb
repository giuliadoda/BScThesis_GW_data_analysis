```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import ScalarFormatter
```

```python
matches = np.loadtxt("matches.txt")
errs = np.loadtxt("matches_errs.txt")
snr = np.loadtxt("SNR.txt")
```

```python
labels = [ 'GW150914 (d = 44 Mpc)',
          'GW150914 (d = 440 Mpc, original)',
        'GW150914 (d = 4400 Mpc)',
        'GW150914 (d = 20 Gpc)',
        'GW150914 (d = 44 Gpc)',
        'GW150914 (d = 100 Gpc)',
        'GW150914 (d = 440 Gpc)',
        'GW150914 (d = 4400 Gpc)',
        'GW150914 (d = 44 Tpc)',
        'GW150914 (d = 440 Tpc)' ]
```

```python
# plot

plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['font.family']      = 'serif'

fig, ax = plt.subplots(figsize=(10,6), constrained_layout=True)

# grid
ax.grid(True, which='major', axis='both', linestyle='--', linewidth=0.5, color='grey', alpha=0.3) 
ax.set_axisbelow(True) 
ax.set_xscale('log')

# markers
markers = [ "o", "v", "<", "D", "s", "P", "X", "d", "^", ">"]

for i in range(0, 10):

    ax.scatter(snr[i], matches[i], color=plt.cm.viridis(i/10), marker="o", label=labels[i], s=120)

# labels
ax.set_xlabel('SNR ', fontdict={'fontfamily':'serif', 'fontsize':18})
ax.set_ylabel('Match', fontdict={'fontfamily':'serif', 'fontsize':18})
ax.set_title('Match vs. SNR - Newtonian chirps', fontdict={'fontfamily':'serif', 'fontsize':18})

# ticks 
ax.tick_params(axis='both', which='both', direction='in', top=True, right=True, size=5, labelsize=16)

# axis limits
ax.set_xlim(3.5*10**(-1), 4*10**3)
ax.set_ylim(-0.12, 1.12)

ax.legend( loc='best', prop={'size': 14})

# Display the plot
plt.show()
```
